```python
import pandas as pd
data0=pd.read_csv("file:///C:/Users/NAAZ%20FATIMA/Downloads/archive/machine_failure_data.csv")
data0.shape
data0.head()
```

```python
a=data0.isnull().sum()/len(data0)*100
#data0.columns
```

```python
data0.head()
```

```python
new=data0.columns
for i in range(24):
    if a[i]>30:
        del data0[new[i]]
data0.head()
```

```python
data0=data0.drop(['Max_Temp','Parameter7_9am','Parameter7_3pm','Parameter4_9am','Parameter4_3pm','Parameter5_3pm','Parameter3_9am','Failure_today'],1)
data0.corr()
```

```python
data0.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
#X_train = l.fit_transform(X_train)
for i in range(12):
    data0.iloc[:,i] = le.fit_transform(data0.iloc[:,i])
```

```python
from sklearn.ensemble import RandomForestRegressor
df=data0.drop('Fail_tomorrow',axis=1)
model=RandomForestRegressor(random_state=1,max_depth=12)
df=pd.get_dummies(df)
model.fit(df,data0.Fail_tomorrow)
```

```python
import numpy as np
import matplotlib.pyplot as plt
features=df.columns
imp=model.feature_importances_
ind=np.argsort(imp)[-11:]
plt.title('feature sort')
plt.barh(range(len(ind)),imp[ind],align='center')
plt.yticks(range(len(ind)),[features[i] for i in ind])
plt.xlabel('rel importance')
plt.show()
```

```python
train_set,test_set=train_test_split(data0,test_size=0.2,random_state=1)
X_train=train_set.iloc[:,8:11].values
Y_train=train_set.iloc[:,11].values
X_test=test_set.iloc[:,8:11].values
Y_test=test_set.iloc[:,11].values
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='rbf',random_state=42)
classifier.fit(X_train,Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
test_set["Predictions"] = Y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_pred)
accuracy = float(cm.diagonal().sum())/len(Y_test)
print("\nAccuracy Of SVM For The Given Dataset : ", accuracy)
```
