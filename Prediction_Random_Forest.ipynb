```python
import pandas as pd
data0=pd.read_csv("file:///C:/Users/NAAZ%20FATIMA/Downloads/archive/machine_failure_data.csv")
#data0.shape
data0.head()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le=LabelEncoder()
col=data0.columns
typ=data0.dtypes
for i in range(len(col)):
    if typ[i]!='object':
        s=col[i]
        data0[s].fillna(data0[s].mean() ,inplace=True)
    else:
        data0.iloc[:,i]=le.fit_transform(data0.iloc[:,i])
data0=pd.get_dummies(data0)
data0.isnull().sum()
```

```python
data0=data0.drop(['RISK_MM'],1)
data0.head()
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(data0,test_size=0.2,random_state=1)
k=22
X_train=train_set.iloc[:,0:k].values
Y_train=train_set.iloc[:,k].values
X_test=test_set.iloc[:,0:k].values
Y_test=test_set.iloc[:,k].values
```

```python
data0['Date'] = pd.to_datetime(data0['Date']).astype('int64')
max_a = data0.Date.max()
min_a = data0.Date.min()
min_norm = -1
max_norm =1
data0['Date'] = (data0.Date- min_a) *(max_norm - min_norm) / (max_a-min_a) + min_norm
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.fit(X_train,Y_train)
```

```python
Y_pred = clf.predict(X_test)
test_set["Predictions"] = Y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_pred)
accuracy = float(cm.diagonal().sum())/len(Y_test)
print("\nAccuracy Of Random Forest Classifier For The Given Dataset : ", accuracy)
```
